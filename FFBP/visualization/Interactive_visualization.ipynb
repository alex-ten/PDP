# Import the required modules

```python
%matplotlib notebook
import time
from collections import OrderedDict
import matplotlib.pyplot as plt
import FFBP.visualization.funcs as vf
from FFBP.visualization.NetworkData import NetworkData
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

# Run visualization
1) NetworkData loads an example pickled snapshot created after testing and instantiates and object

2) vf.prep_figure() prepares a figure of the required size and returns variables for vf.prep_all_axes() function

3) vf.prep_all_axes() places mpl.plt.axes() objects and text label inside each

4) interact() takes vf.draw_all_layers() to interactively feed values returned by w_ep and w_pat widgets into it

```python
net_data = NetworkData('example1.pkl')

logs = net_data.main.values()

fig, figure_map = vf.prep_figure(net_data, cpi = 10)

axs = vf.prep_all_axes(fig, figure_map, net_data)

# Test Values
vals = [0,1,2,3]
labs = ['p1','p2','p3','p4']
dropDownDict = OrderedDict(zip(labs,vals))

w_ep = widgets.IntSlider(value=0, min=0, max=10,step=1, description='epoch')
w_pat = widgets.Dropdown(options=dropDownDict, value=0, description='pattern')

interact(vf.draw_all_layers, 
         fig=fixed(fig), 
         logs=fixed(logs),
         axes=fixed(axs),
         epoch = w_ep,
         pattern = w_pat,
         __manual=True,
        )
```
